**General Plot.py**

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#10 x 1 Matrix
x=np.random.random((10,1))
print(x)
```

```
plt.plot(x, '*-') #We may give '+' instead of Star
plt.show()
#Y-axis - Data and x-axis - instance of the Data
#Starts with 0 to 9
```

```
#Data starts from 0 to 100 , atleast 50 datas must be present inside this range.
x=np.linspace(0,100,50)
y=np.power(x,.5) #Square root of X data
z=np.power(x,2) #Square  of X data
plt.plot(x,y,'+-')
plt.show()
plt.plot(x,z,'*-')
plt.show()
 
#if we give only one plot.show() at the last then 
    #two function graph will appear in one sheet
```

```
#Plot the same graph using Seaborn
sns.set()
sns.lineplot(x,y)
plt.show()
```
